```python
import pandas as pd
df = pd.read_csv('data/weather_data_5.csv')
df
```

```python
# parsing the day column as "DATA" type 
df = pd.read_csv('data/weather_data_5.csv',parse_dates=["day"])
df
```

```python
# set index to day
df.set_index('day',inplace=True)
df
```

```python
# filling the NaN values
n_df = df.fillna(0)
n_df
```

```python
# fillna with custom columns
n_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
n_df
```

```python
# carry forward the previous value
n_df = df.fillna(method='ffill')
n_df
```

```python
# carry backward the next value
n_df = df.fillna(method='bfill')
n_df
```

```python
# adding the limit to copy the value
n_df = df.fillna(method='ffill', limit=1)
n_df
```
