```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('data/homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price,color='red', marker='*')
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.plot(df.area, df.price, linestyle='solid')
```

```python
# jo predict kar na hai usko bahar me likh te hai 
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
x = reg.predict([[3300]])
nd = {'price':x[0],'area':'3300'}
df.append(nd,ignore_index = True)
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price,color='red', marker='*')
```

```python
new_df = pd.read_csv('data/areas.csv')
new_df
```

```python
price = reg.predict(new_df)
```

```python
new_df['price'] = price
new_df
```

```python
new_df.to_csv('data/areas_price.csv',index=False) 
```
