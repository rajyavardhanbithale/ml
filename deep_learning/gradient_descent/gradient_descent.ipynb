```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
x = df[['age','affordibility']]
y = df[['bought_insurance']]
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['age','affordibility']],df.bought_insurance,test_size=0.2, random_state=25)
```

```python
x_train.shape
```

```python
x_train_scale = x_train.copy()
x_train_scale.age = x_train_scale.age/100

x_test_scale = x_test.copy()
x_test_scale.age = x_test_scale.age/100
```

```python
len(x_train)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train_scale,y_train,epochs=1300)
```

model.evaluate(x_test_scale,y_test)
