```python
import pandas as pd
```

```python
df = pd.read_csv('data/stock_data.csv')
df
```

```python
# removing messed up column using "skiprows"
df = pd.read_csv('data/stock_data.csv', skiprows=1)
df
```

```python
# removing messed up column using "header"
df = pd.read_csv('data/stock_data.csv', header=1)
df
```

```python
# custom header to the data [index]
df = pd.read_csv('data/stock_data_no_header.csv',header=None)
df
```

```python
# custom header to the data [custom name to columns]
df = pd.read_csv('data/stock_data_no_header.csv', header=None, names=['tickers','eps','revenue','price','people'])
df
```

```python
# limit the rows
df = pd.read_csv('data/stock_data_new.csv', nrows=2)
df
```

```python
# cleaning up the messy data
# replacing the messy value with NaN
df = pd.read_csv('data/stock_data_new.csv', na_values=['not available','n.a.'])
df
```

```python
# replacing the messy value with NaN 
# with custom selected columns
df = pd.read_csv('data/stock_data_new.csv', na_values={
    'eps': ['not available','n.a.'],
    'revenue': ['not available','n.a.',-1],
    'price':['not available','n.a.'],
    'people':['not available','n.a.']
})
df
```

```python
# writing back to csv
df.to_csv('data/stock_data_save.csv')
```

```python
# writing back to csv without index
df.to_csv('data/stock_data_save_wo_index.csv',index=False)
```
